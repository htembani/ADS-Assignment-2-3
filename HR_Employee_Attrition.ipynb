```python
import pandas as pd
import numpy as np; np.random.seed(42)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data = pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv") 
df = Data.copy()
```

```python
#All numeric fields being converted to categoric
Data['Age'] = Data.Age.astype('category')
Data['DailyRate'] = Data.DailyRate.astype('category')
Data['DistanceFromHome'] = Data.DistanceFromHome.astype('category')
Data['Education'] = Data.Education.astype('category')
Data['EmployeeCount'] = Data.EmployeeCount.astype('category')
Data['EmployeeNumber'] = Data.EmployeeNumber.astype('category')
Data['EnvironmentSatisfaction'] = Data.EnvironmentSatisfaction.astype('category')
Data['HourlyRate'] = Data.HourlyRate.astype('category')
Data['JobInvolvement'] = Data.JobInvolvement.astype('category')
Data['NumCompaniesWorked'] = Data.NumCompaniesWorked.astype('category')
Data['PercentSalaryHike'] = Data.PercentSalaryHike.astype('category')
Data['PerformanceRating'] = Data.PerformanceRating.astype('category')
Data['RelationshipSatisfaction'] = Data.RelationshipSatisfaction.astype('category')
Data['StandardHours'] = Data.StandardHours.astype('category')
Data['StockOptionLevel'] = Data.StockOptionLevel.astype('category')
Data['TotalWorkingYears '] = Data.TotalWorkingYears .astype('category')
Data['TrainingTimesLastYear'] = Data.TrainingTimesLastYear.astype('category')
Data['WorkLifeBalance'] = Data.WorkLifeBalance.astype('category')
Data['YearsAtCompany'] = Data.YearsAtCompany.astype('category')
Data['YearsInCurrentRole'] = Data.YearsInCurrentRole.astype('category')
Data['YearsSinceLastPromotion'] = Data.YearsSinceLastPromotion.astype('category')
Data['YearsWithCurrManager'] = Data.YearsWithCurrManager.astype('category')
```

```python
Data.head()#checking firstdataframe structure after the conversion
```

```python
# Plot below shows that daily rate is highly dense between ages 25 to 40, meaning that people are very active in that range 
fig, ax = plt.subplots(figsize=(16, 6))
ax.scatter(x= Data['Age'], y= Data['DailyRate'],alpha = 0.5,color = 'red')
plt.xlabel('Age', fontsize=18)              # label = name of label
plt.ylabel('Daily Rate', fontsize=18)
plt.title('Age vs DailyRate Scatter Plot', fontsize=20)            # giving the plot an appropriate tittle
```

```python
# Histogram below shows that TotalWorkingYears highest in 5 to 10 Years 
# bins represent number of bar in figure
Data.TotalWorkingYears.plot(kind = 'hist',bins = 10,figsize = (15,15))
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x="JobRole", y="DistanceFromHome", hue="Attrition",ci = None, data= df);#confidence level (ci) set to none
plt.xticks(rotation = 45, ha = 'right')
plt.grid()  #inserting grid lines in our graphh
plt.xlabel('Job Role', fontsize = 15, weight = 'bold')
plt.ylabel('Distsnce From Home', fontsize = 15, weight = 'bold')
plt.title('Employee Attrition by Job Role and Distance from Home', fontsize = 15, weight = 'bold')
plt.show()
```

```python
sns.boxplot(Data['Gender'], Data['MonthlyIncome'])
plt.title('MonthlyIncome vs Gender Box Plot', fontsize=20)      
plt.xlabel('MonthlyIncome', fontsize=16)
plt.ylabel('Gender', fontsize=16)
plt.show()

## Inference
# Average and range of salary is higher in Female compared to Male associates
```

```python
#Count Plot
plt.figure(figsize=(16, 5))
sns.countplot(Data.JobLevel)
plt.title('JobLevel Count Plot', fontsize=20)      
plt.xlabel('Job Level', fontsize=16)
plt.ylabel('Count', fontsize=16)
plt.show()

## Inference
# Largest population is between Joblevel 1 and 2
```

```python
# BarPlot
plt.figure(figsize=(15, 5))
sns.barplot(x ='Attrition', y = 'DistanceFromHome', data = df)
plt.xlabel('Attrition', fontsize = 15, weight = 'bold')
plt.ylabel('Distsnce From Home', fontsize = 15, weight = 'bold')
plt.title('Employee Attrition Vs Distance from Home', fontsize = 15, weight = 'bold')
## Inference: 
# People are more likely to quit When 'DistanceFromHome' is between 9 to 12
```

```python
#sns.distplot(data['Age']) showing that there is a normal distribution of Age and Mean age is about 35 Years
plt.figure(figsize=(15, 5))
sns.distplot(Data.Age, kde=True, label='YearsInCurrentRole', hist_kws={"histtype": "step", "linewidth": 3,
                  "alpha": 1, "color": sns.xkcd_rgb["azure"]})

plt.title('Age Distribution plot', fontsize=20)      
plt.show()
```

```python
# pie chart of workers shows that there are more males than females
labels = ['Male', 'Female']
sizes = [Data['Gender'].value_counts()[0],
         Data['Gender'].value_counts()[1]
        ]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
ax1.axis('equal')
plt.title('Gender Pie Chart', fontsize=20)
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
sns.distplot(Data.YearsWithCurrManager, hist=False, kde=True, label='YearsWithCurrManager', hist_kws={"histtype": "step", "linewidth": 3,
                  "alpha": 1, "color": sns.xkcd_rgb["azure"]})
sns.distplot(Data.YearsAtCompany, hist=False, kde=True, label='YearsAtCompany', hist_kws={"histtype": "step", "linewidth": 3,
                  "alpha": 1, "color": sns.xkcd_rgb["dark blue green"]})
sns.distplot(Data.YearsInCurrentRole, hist=False, kde=True, label='YearsInCurrentRole', hist_kws={"histtype": "step", "linewidth": 3,
                  "alpha": 1, "color": sns.xkcd_rgb["fuchsia"]})
plt.suptitle('Total Years: Current Manager, Role and At Company', size=22, x=0.5, y=0.94)
plt.xlabel('Years', size=16)
plt.ylabel('Count', size=16)
plt.legend(prop={'size':26}, loc=1)
plt.show();
```

```python
plt.figure(figsize=(15, 5))
sns.pointplot(x="Gender", y="TotalWorkingYears", hue="Attrition", data=Data,
              palette={"Yes": "blue", "No": "pink"},
              markers=["*", "o"], linestyles=["-", "--"])
plt.xlabel('Gender', fontsize = 15, weight = 'bold')
plt.ylabel('Total Working Years', fontsize = 15, weight = 'bold')
plt.title('Total Working Years Vs. Attrition', fontsize = 15, weight = 'bold');

## Inference
# Males and Females within the ranges for 'TotalWorkingYears' of 11 to 13 are less likely to quit
```

```python
cont_col= ['DistanceFromHome', 'PerformanceRating','EnvironmentSatisfaction','Attrition','YearsWithCurrManager','NumCompaniesWorked']
sns.pairplot(df[cont_col], kind="reg", diag_kind = "kde" , hue = 'Attrition' )
plt.show();

## Inference - Employees falling under the below categories are more likely to quit
# 'PerformanceRating' between 3.0 & 4.0 and 'DistanceFromHome' between 10 & 15
# 'YearsWithCurrManager' less than 5
```
